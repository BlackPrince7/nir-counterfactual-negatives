````python
import torch
import json
import numpy as np
from transformers import AutoModelForCausalLM, AutoTokenizer
from sentence_transformers import SentenceTransformer, util
from rank_bm25 import BM25Okapi
import re
import ast

embed_model = SentenceTransformer("BAAI/bge-base-en-v1.5", device="cuda")

llm_id = "Qwen/Qwen2.5-3B-Instruct"
llm_tokenizer = AutoTokenizer.from_pretrained(llm_id)
llm_model = AutoModelForCausalLM.from_pretrained(
    llm_id, 
    device_map="cuda", 
    load_in_4bit=True
)

def get_similarity(q, d):
    """Вычисляет косинусное сходство между текстами"""
    emb_q = embed_model.encode(q, convert_to_tensor=True)
    emb_d = embed_model.encode(d, convert_to_tensor=True)
    return util.cos_sim(emb_q, emb_d).item()

def generate_llm_response(prompt, max_new_tokens=256):
    """Обертка для вызова LLM"""
    messages = [{"role": "user", "content": prompt}]
    text = llm_tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    inputs = llm_tokenizer([text], return_tensors="pt").to("cuda")
    
    with torch.no_grad():
        outputs = llm_model.generate(**inputs, max_new_tokens=max_new_tokens, temperature=0.7)
    
    return llm_tokenizer.decode(outputs[0][inputs.input_ids.shape[1]:], skip_special_tokens=True)

def extract_facts(document):
    """
    Извлечение фактов из текста с надежным парсингом JSON-ответа LLM.
    Обрабатывает Markdown-блоки, мусорный текст и мелкие синтаксические ошибки.
    """
    prompt = (
        "Extract key facts from the text strictly as a JSON array of arrays. "
        "Do not include any explanations, greetings, or markdown code blocks. "
        "Format must be exactly: [[\"subject\", \"action\", \"result/number\"]]\n"
        f"Text: {document}"
    )
    
    response = generate_llm_response(prompt).strip()
    
    cleaned = re.sub(r'^```(?:json)?|```$', '', response, flags=re.MULTILINE).strip()
    
    match = re.search(r'\[\s*\[.*\]\s*\]', cleaned, re.DOTALL)
    if not match:
        return []
        
    json_str = match.group(0)

    try:
        parsed = json.loads(json_str)
        if isinstance(parsed, list) and all(isinstance(i, list) for i in parsed):
            return parsed
    except json.JSONDecodeError:
        try:

            parsed = ast.literal_eval(json_str)
            if isinstance(parsed, list) and all(isinstance(i, list) for i in parsed):
                return parsed
        except (ValueError, SyntaxError):
            pass

    return []

def mutate_fact(document, query, fact):
    """Генерация контрфактического документа d⁻"""
    prompt = f"Context: {document}\nChange ONLY the fact {fact} so the new text is a factually INCORRECT answer to the query: '{query}'. Do not change the style or structure."
    return generate_llm_response(prompt)

corpus_tokenized = [doc.split() for doc in all_documents]
bm25_index = BM25Okapi(corpus_tokenized)

def check_bm25(query, d_neg, threshold=1.0):
    score = bm25_index.get_scores(query.split())[corpus.index(d_neg)]
    return score < threshold
````
